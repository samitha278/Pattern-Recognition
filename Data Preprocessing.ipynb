## Missing Value treatments

```python
#imputing missing value methods : mean median mode KNNIputer

#life expecteny is target variable not doing any treatment

for i in [" BMI ","Polio","Income composition of resources"]:
    k = df[i].median()
    df[i].fillna(k)
    
```

```python
df.isnull().sum()    #after imput
```

```python
df.isnull().sum()    #before imput
```

```python
from sklearn.impute import KNNImputer

impute = KNNImputer()


for i in df.select_dtypes(include="number").columns:
    df[i]=impute.fit_transform(df[[i]])
```
